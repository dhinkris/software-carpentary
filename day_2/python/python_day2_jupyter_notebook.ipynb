```python
def get_urban(df,threshold):
    urban = df.loc[df["Population_Density"]>=threshold,:]
    rural = df.loc[df["Population_Density"]<threshold,:]
    
    return [urban,rural]
```

```python
get_urban(dem_health_data,10000)[0].shape
```
